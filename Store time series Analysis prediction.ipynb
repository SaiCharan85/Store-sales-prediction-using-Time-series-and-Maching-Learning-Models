```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Resample the data to weekly frequency and sum the weekly sales
weekly_sales = df['Weekly_Sales'].resample('W').sum()



# Plot weekly sales over time
plt.figure(figsize=(12, 6))
plt.plot(weekly_sales)
plt.title('Weekly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.show()



```

```python
# Decompose the time series into trend, seasonal, and residual components

decomposition = seasonal_decompose(weekly_sales, model='additive')
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot the decomposed components
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(weekly_sales, label='Original', color='blue')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend, label='Trend', color='blue')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality', color='blue')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual, label='Residuals', color='blue')
plt.legend(loc='upper left')
plt.tight_layout()

# Autocorrelation plot
plt.figure(figsize=(12, 6))
plot_acf(weekly_sales, lags=52, ax=plt.gca())
plt.show()

# Partial autocorrelation plot
plt.figure(figsize=(12, 6))
plot_pacf(weekly_sales, lags=52, ax=plt.gca())
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
X= df_dum.drop(["Weekly_Sales","Store","Dept","Date"],axis=1)
y= df_dum["Weekly_Sales"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score, classification_report
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
XGB = XGBRegressor()
XGB.fit(x_train, y_train)
y_pred = XGB.predict(x_test)
print("Mean Absolute Error XGB model: " + str(mean_absolute_error(y_pred, y_test)))
```

```python
XGB = XGBRegressor(n_estimators=7, max_depth = None, learning_rate =0.1)
XGB.fit(x_train, y_train, 
         early_stopping_rounds=5, 
         eval_set=[(x_test, y_test)],
         verbose=False
       )
y_pred = XGB.predict(x_test)
print("Mean Absolute Error XGB model: " + str(mean_absolute_error(y_pred, y_test)))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual Values', color = "blue")
plt.ylabel('Predicted Values', color = "green")
plt.title('XGB Predictions')
plt.show()
```

```python
RFR = RandomForestRegressor()
RFR.fit(x_train, y_train)
y_pred1 = RFR.predict(x_test)
print("Mean Absolute Error RandomForest: " + str(mean_absolute_error(y_pred1, y_test)))
```

```python
RFR = RandomForestRegressor(random_state = 0, max_depth = None, min_samples_leaf = 1,
min_samples_split= 2,
n_estimators = 200)
RFR.fit(x_train, y_train)
y_pred1 = RFR.predict(x_test)
print("Mean Absolute Error RandomForest: " + str(mean_absolute_error(y_pred1, y_test)))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, y_pred1, alpha=0.5)
plt.xlabel('Actual Values', color = "blue")
plt.ylabel('Predicted Values', color = "green")
plt.title('RF Predictions')
plt.show()
```

```python
LR = LinearRegression()
LR.fit(x_train, y_train)
y_pred2 = LR.predict(x_test)
print("Mean Absolute Error RandomForest: " + str(mean_absolute_error(y_pred2, y_test)))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, y_pred2, alpha=0.5)
plt.xlabel('Actual Values', color = "blue")
plt.ylabel('Predicted Values', color = "green")
plt.title('Linear Regression Predictions')
plt.show()
```

XG Boost is chosen as the primary model for our time series forecasting of the sales as the MAE is the least compared to other models
